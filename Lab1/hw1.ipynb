```python
import numpy as np
import matplotlib.pyplot as plt

# Define the discrete-time signal x[n]
def x(n):
    if n == -4:
        return -1
    elif n == -3:
        return -0.5
    elif n == -2:
        return 0.5
    elif -1 <= n <= 2:
        return 1
    elif n == 3:
        return 0.5
    else:
        return 0

n_values_a = np.arange(-8, 15) 
x_3_minus_n = [x(3 - n) for n in n_values_a]

n_values_b = np.arange(-8, 15)
x_n_minus_1_squared = [x((n - 1)**2) for n in n_values_b]
```

```python
# Plot part (a) without 'use_line_collection'
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.stem(n_values_a, x_3_minus_n, basefmt=" ")
plt.title(r"$x[3 - n]$")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)

# Plot part (b) without 'use_line_collection'
plt.subplot(1, 2, 2)
plt.stem(n_values_b, x_n_minus_1_squared, basefmt=" ")
plt.title(r"$x[(n - 1)^2]$")
plt.xlabel("n")
plt.ylabel("x[n]")
plt.grid(True)

plt.tight_layout()
plt.show()
```
